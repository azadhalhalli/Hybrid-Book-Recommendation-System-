```python
import pandas as pd
import numpy as np

# Sayısal ve boş değerleri temizleme
df_cleaned = df[df['country'].apply(lambda x: isinstance(x, str))]

# Virgül içeren satırları kaldırma
df_cleaned = df_cleaned[~df_cleaned['country'].str.contains(',')]

# Ülke isimlerini küçük harfe çevirip boşlukları temizleme
df_cleaned['country'] = df_cleaned['country'].str.strip().str.lower()

# Temizlenmiş ülke isimlerini kontrol et
df_cleaned['country'].value_counts()
```

```python
# NaN ve boş değerleri temizle
df_cleaned = df.dropna(subset=['country'])

# Virgül içeren satırları kaldırma ve yalnızca string olanlara uygula
df_cleaned = df_cleaned[df_cleaned['country'].apply(lambda x: isinstance(x, str) and ',' not in x)]

# Ülke isimlerini küçük harfe çevirip boşlukları temizleme
df_cleaned['country'] = df_cleaned['country'].str.strip().str.lower()

# Temizlenmiş ülke isimlerini kontrol et
df_cleaned['country'].value_counts()
```

```python
country_counts = df_cleaned['country'].value_counts().reset_index()
country_counts.columns = ['country', 'count']
country_counts['percentage'] = (country_counts['count'] / df_cleaned['country'].count()) * 100
country_counts
```

```python
# Amerika için farklı yazım şekillerini tek tipe dönüştürme (df_cleaned kullanarak)
df_cleaned['country'] = df_cleaned['country'].replace({
    'united states of america': 'usa',
    'u.s. of a.': 'usa',
    'united states': 'usa',
    'america': 'usa'
    
})

# Dönüştürülen veri setini kontrol et
df_cleaned['country'].value_counts()
```

```python
country_counts = df_cleaned['book_author'].value_counts().reset_index()
country_counts.columns = ['book_author', 'count']
country_counts['percentage'] = (country_counts['count'] / df_cleaned['book_author'].count()) * 100
country_counts
```

```python
country_counts = df_cleaned['city'].value_counts().reset_index()
country_counts.columns = ['city', 'count']
country_counts['percentage'] = (country_counts['count'] / df_cleaned['city'].count()) * 100
country_counts
```

```python
country_counts = df_cleaned['Category'].value_counts().reset_index()
country_counts.columns = ['Category', 'count']
country_counts['percentage'] = (country_counts['count'] / df_cleaned['Category'].count()) * 100
country_counts
```

```python
df1 = pd.DataFrame(df_cleaned)

# Köşeli parantez ve tırnak işaretlerini kaldırma
df1['Category'] = df1['Category'].str.replace("[\[\]']", "", regex=True)

# Parantez içindeki verileri ayırma ve ilk türü seçme
df1['Category'] = df1['Category'].apply(lambda x: x.split(',')[0].split('(')[0].strip())

# Temizlenmiş Category sütununu kontrol et
df1['Category'].value_counts()
```

```python
country_counts = df1['Category'].value_counts().reset_index()
country_counts.columns = ['Category', 'count']
country_counts['percentage'] = (country_counts['count'] / df1['Category'].count()) * 100
country_counts
```

```python
# Temizleme işlemleri
df2 = df1.copy()

# Tırnak işaretlerini kaldırma
df2['Category'] = df2['Category'].str.replace('"', '', regex=False)

# `&` işaretlerini kaldırma ve ikili türleri düzeltme
df2['Category'] = df2['Category'].str.replace('&', ',', regex=False)

# Her kategoriye virgüller ile ayrılmış türler varsa sadece ilkini seçme
df2['Category'] = df2['Category'].apply(lambda x: x.split(',')[0].strip())

# df1 ve df2'yi kontrol et
print("df1:")
print(df1.head())
print("\ndf2:")
print(df2.head())
```

```python
country_counts = df2['Category'].value_counts().reset_index()
country_counts.columns = ['Category', 'count']
country_counts['percentage'] = (country_counts['count'] / df2['Category'].count()) * 100
country_counts
```

```python
category_freq = df2['Category'].value_counts().to_dict()
df2['Category_frequency'] = df2['Category'].map(category_freq)
```

```python
language_freq = df2['Language'].value_counts().to_dict()
df2['Language_frequency'] = df2['Language'].map(language_freq)
```

```python
country_freq = df2['country'].value_counts().to_dict()
df2['Country_frequency'] = df2['country'].map(country_freq)
```

```python
city_freq = df2['city'].value_counts().to_dict()
df2['City_frequency'] = df2['city'].map(city_freq)
```

```python
publisher_freq = df2['publisher'].value_counts().to_dict()
df2['Publisher_frequency'] = df2['publisher'].map(publisher_freq)
```
